---

## üóÇÔ∏è Types of Operators

### üîó Logical Operators
Logical operators connect two or more conditions and evaluate the truth or falsity of the conditions combined.

| Operator | Name | Example | Result |
|----------|------|---------|---------|
| `and` | Logical AND | `x and y` | ‚úÖ True if both are True |
| `or` | Logical OR | `x or y` | ‚úÖ True if at least one is True |
| `not` | Logical NOT | `not x` | ‚úÖ Reverse the result |

### üÜî Identity Operators
Identity operators compare the memory locations of two objects.

| Operator | Name | Example | Result |
|----------|------|---------|---------|
| `is` | Identity | `x is y` | ‚úÖ True if same object |
| `is not` | Non-identity | `x is not y` | ‚úÖ True if different objects |

### üîç Membership Operators
Membership operators test if a sequence contains a specific value.

| Operator | Name | Example | Result |
|----------|------|---------|---------|
| `in` | Membership | `x in y` | ‚úÖ True if value found |
| `not in` | Non-membership | `x not in y` | ‚úÖ True if value not found |

### üíª Bitwise Operators
Bitwise operators work on bits and perform bit-by-bit operations.

| Operator | Name | Example | Result |
|----------|------|---------|---------|
| `&` | AND | `x & y` | Sets each bit to 1 if both bits are 1 |
| `\|` | OR | `x \| y` | Sets each bit to 1 if any bit is 1 |
| `^` | XOR | `x ^ y` | Sets each bit to 1 if only one bit is 1 |
| `~` | NOT | `~x` | Inverts all bits |
| `<<` | Left shift | `x << 2` | Shift left by pushing zeros in from right |
| `>>` | Right shift | `x >> 2` | Shift right by pushing copies from left |

## Identity Operators

Identity operators check if two variables occupy the same memory location, i.e., are identical.

| Operator | Name          | Example       |
|-----------|---------------|----------------|
| `is`      | Identity      | `x is y`       |
| `is not`  | Not identity  | `x is not y`   |

> ‚ö†Ô∏è This is **NOT** checking if something is equal to something else.


```python
salary_haroun = 100_000
salary_abdo = 100_000

salary_haroun is salary_abdo
```

```python
haroun_skills = ['Python' , 'R' , 'SQL']
abdo_skills = haroun_skills

abdo_skills is haroun_skills
```

## Membership Operators

Membership operators test for membership in a sequence, such as strings, lists, or tuples.

| Operator | Name             | Example       |
|-----------|------------------|----------------|
| `in`      | Membership       | `x in y`       |
| `not in`  | Not membership   | `x not in y`   |

## Bitwise Operators

Bitwise operators perform bit-by-bit operations on binary numbers. üíª

| Operator | Name          | Example    |
|-----------|---------------|-------------|
| `&`       | Bitwise AND   | `x & y`     |
| `|`       | Bitwise OR    | `x | y`     |
| `^`       | Bitwise XOR   | `x ^ y`     |
| `~`       | Bitwise NOT   | `~x`        |
| `<<`      | Left Shift    | `x << y`    |
| `>>`      | Right Shift   | `x >> y`    |

### NOTE:
The technical use case is for bit-wise operation (shown below).

```python
x = 10  # binary 1010
y = 4   # binary 0100
result = x & y  # binary 0000 -> decimal 0

result
```

```python
x = 10  # binary 1010
y = 4   # binary 0100
result = x | y  # binary 1110 -> decimal 14

result
```
