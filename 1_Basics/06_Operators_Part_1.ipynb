# Order of Operations

## ğŸ¯ **Hierarchy of Operations**

### 1. ğŸ“ **Parentheses** 
- ğŸª Anything inside parentheses `( )`, brackets `[ ]`, or braces `{ }` is executed first
- **Example**: `(3 + 2) Ã— 4` â†’ `5 Ã— 4 = 20`
- âœ… **First Priority**: Always start with expressions inside grouping symbols

### 2. ğŸš€ **Exponents**
- âš¡ Operations involving:
  - ğŸ”¢ Exponents: `xÂ²`, `yÂ³`
  - ğŸŒ± Roots: `âˆšx` (equivalent to `x^(1/2)`)
  - ğŸ“Š Fractional exponents: `x^(a/b)`
- **Example**: `2Â³ + âˆš9` â†’ `8 + 3 = 11`
- âœ… **Second Priority**: Handle all exponential operations next

### 3. âœ–ï¸â— **Multiplication and Division**
- ğŸ”„ Processed from **left to right**
- âš–ï¸ Equal priority - neither takes precedence over the other
- **Example**: `12 Ã· 3 Ã— 2` â†’ `4 Ã— 2 = 8`
- âœ… **Third Priority**: Work through multiplication and division as they appear

### 4. â•â– **Addition and Subtraction**
- ğŸ”„ Processed from **left to right**
- âš–ï¸ Equal priority - neither takes precedence over the other
- **Example**: `10 - 3 + 2` â†’ `7 + 2 = 9`
- âœ… **Fourth Priority**: Finally, handle addition and subtraction
