`if`, `elif`, and `else` are conditional statements that control the flow of execution based on whether a condition is `True` or `False`.

# Conditional Statements

| Part | Icon | Purpose | Condition Evaluated | Action if True |
|------|------|---------|-------------------|----------------|
| `if` | ðŸ” | First condition check | First condition to be evaluated | Executes block if condition is true |
| `elif` | ðŸ”„ | Subsequent condition check | Next condition, if `if` is false | Executes block if this condition is true |
| `else` | âš¡ | Fallback if no conditions are met | None | Executes block if all above conditions are false |

## ðŸŽ¯ **Syntax Structure**
```python
if condition1:
    # âœ… Executes if condition1 is True
elif condition2:
    # ðŸ”„ Executes if condition1 is False and condition2 is True  
elif condition3:
    # ðŸ”„ Executes if all above are False and condition3 is True
else:
    # âš¡ Executes if all conditions are False

````python
if True:
    print("What's up?")
````

Here's another example that if the first condition is False. Then let's do something else.

````python
if False:
    print("Hey Data Nerd")
else:
    print("What's up Dude")
````

# If Statement ðŸ”

## ðŸ“ **Notes**
- `if`: Evaluates a condition âš–ï¸
- âœ… If the condition is **true**, the block of code under the `if` statement executes
- âŒ If the condition is **false**, it skips this block

## ðŸ’¼ **Example**
Let's see if our skill matches with a job's required skills. We'll use variables to represent the applicant's skill and the skill required for the job, then compare them using an `if` statement.

### ðŸŽ¯ **Variable Definitions**
```python
applicant_skill = "Python"        # ðŸ Skill the applicant has
job_skill = "Python"              # ðŸ’¼ Skill required for the job posting
result = "No suitable job found"  # ðŸ” Initial result

````python
# Applicant Skill
applicant_skill = 'SQL'

# Skill required for job posting
job_skill = 'SQL'

result = 'No Skills Match'
````

````python
if applicant_skill == job_skill:
    result = 'Skills Match'
else:
    result = 'No Skills Match'
result
````

# Elif Statement ðŸ”„

## ðŸ“ **Notes**
- `elif`: Follows an `if` or another `elif` statement
- ðŸ”„ It means "if the previous conditions were not true, then try this condition"
- âœ… If the `elif` condition is true, its code block executes

## ðŸ’¼ **Example**
Now we're adding a variable for years of experience to determine job level.

### ðŸŽ¯ **Variable Definitions**
```python
applicant_skill = "Python"        # ðŸ Skill the applicant has
job_skill = "Python"              # ðŸ’¼ Skill required for the job posting
years_experience = 3              # ðŸ“… Years of experience applicant has
result = "No suitable job found"  # ðŸ” Initial result

````python
years_experience = 2
# Check if your desired job skill matches the required skill and if you are qualified for the position. 
if applicant_skill == job_skill and (years_experience >= 2):
    result = 'Skill & Experience Match'
else:
    result = 'No Match Found'

# Print the result to inform you of the outcome of your job search
result
````

````python
# Check if the applicant's skill matches the required skill of the job posting and they meet the experience requirement. 
if applicant_skill != job_skill:
    result = 'Skills still needed'
elif applicant_skill== job_skill and years_experience < 2:
    result = 'More experience needed'
else:
    result = 'Experience matches!'

# Print the result
result
````
